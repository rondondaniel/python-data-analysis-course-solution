```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

Nous chargeons cet ensemble de données directement à partir d'une URL. Le TabularDataset d'AutoGluon est une sous-classe de pandas DataFrame, de sorte que toutes les méthodes DataFrame peuvent être utilisées sur TabularDataset.

```python
data_root = 'https://autogluon.s3.amazonaws.com/datasets/Inc/'
```

```python
train_data = TabularDataset(data_root + 'train.csv')
test_data = TabularDataset(data_root + 'test.csv')
```

Nous construisons maintenant un TabularPredictor en spécifiant le nom de la colonne label et nous nous entraînons sur l'ensemble de données avec TabularPredictor.fit().

```python
predictor = TabularPredictor(label='class').fit(train_data=train_data)
```

Une fois que nous avons un prédicteur adapté à l'ensemble de données d'entraînement, nous pouvons charger un ensemble de données distinct à utiliser pour la prédiction et l'évaluation.

```python
predictions = predictor.predict(test_data)
```

Nous pouvons évaluer le prédicteur sur l'ensemble de données de test à l'aide de la fonction evaluate(), qui mesure les performances de notre prédicteur sur des données qui n'ont pas été utilisées pour l'ajustement des modèles.

```python
predictor.evaluate(test_data, silent=True)
```

Le TabularPredictor d'AutoGluon fournit également la fonction leaderboard(), qui nous permet d'évaluer les performances de chaque modèle formé sur les données de test.

```python
predictor.leaderboard(test_data)
```
